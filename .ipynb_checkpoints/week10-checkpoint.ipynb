## Cohort Problems
### Problem 1

```python
import numpy as np
from sklearn.metrics import confusion_matrix
def get_metrics(actual_targets, predicted_targets, labels):
    dictionary = {}
    c_matrix = confusion_matrix(actual_targets, predicted_targets,labels)
    dictionary['confusion matrix'] = c_matrix
    dictionary['total records'] = np.sum(dictionary['confusion matrix'])
    dictionary['accuracy'] = (c_matrix[0,0] + c_matrix[1,1])/np.sum(c_matrix)
    dictionary['sensitivity'] = c_matrix[1,1]/np.sum(c_matrix[[1],:])
    dictionary['false positive rate'] = c_matrix[0,1]/np.sum(c_matrix[[0],:])
    return dictionary

actual = ['cat', 'cat', 'cat', 'cat', 'bird', 'bird','bird','bird']
predicted = ['cat', 'cat', 'bird', 'bird', 'cat', 'bird', 'bird', 'bird']
labels = ['bird', 'cat']
print(get_metrics(actual, predicted, labels))
```

### Problem 2

```python
def five_number_summary(x):
    summary={}
    summary['minimum'] = np.min(x)
    summary['first quartile'] = np.percentile(x,25)
    summary['median'] = np.percentile(x,50)
    summary['third quartile'] = np.percentile(x,75)
    summary['maximum'] = np.max(x)
    return summary

first_column = bunchobject.data[:,[1]]
print(five_number_summary(first_column))
```

```python
col_no = [0,1,2]
some_columns = bunchobject.data[:,col_no]
print( five_number_summary(some_columns) )
```

```python
def normalize_minmax(data):
    return (data-np.min(data))/(np.max(data)-np.min(data))
```

```python
first_column = bunchobject.data[:,[1]]
first_column_norm = normalize_minmax(first_column)
print(five_number_summary(first_column_norm))
```

```python
cols = [1, 7]
some_columns = bunchobject.data[:,cols]
snorm = normalize_minmax(some_columns)
print('normalized', five_number_summary(snorm))
```
